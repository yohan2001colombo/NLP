```python
import re
from nltk.corpus import stopwords
import nltk

raw_text = "This is a sample inputtext! It includes punctuation, some Stopwords, and more."
tokens = re.findall(r'\b\w+\b', raw_text.lower()) # extract all the words without space,underscore..

filtered_words = [
    word for word in tokens
    if word.isalpha() and word not in stopwords.words('english')
]

vocab = sorted(set(filtered_words))
print("\n First 50 uniques filtered words: ")
print(vocab[:50])
```

```python
sample_tweet = "This is an example tweet! #NLP @UCSC ðŸ˜Š"

pattern = r'\b\w+'
regex_tokens = regexp_tokenize(sample_tweet, pattern)
print("\n Regex Tokens:", regex_tokens)

tweet_tokenizer = TweetTokenizer()
tweet_tokens = tweet_tokenizer.tokenize(sample_tweet)
print("Tweet tokens:", tweet_tokens)
```

```python
from nltk.corpus import gutenberg

print(gutenberg.fileids())

file_id = gutenberg.fileids() [0]
raw_text = gutenberg.raw(file_id)

tokens = nltk.word_tokenize(raw_text)
words = [word.lower() for word in tokens]

vocab = sorted(set(words))
print(vocab[:50])
```

```python
from nltk.tokenize import regexp_tokenize

pattern = r'\w+'
regex_tokens = regexp_tokenize(raw_text, pattern)
print("Regex tokens sample:", regex_tokens[:20])
```

```python
from nltk import FreqDist

fdist = FreqDist(words)

fdist.plot(30, cumulative = False)
plt.show()
```

```python
from wordcloud import WordCloud

wordcloud = WordCloud(width = 800, height = 400, background_color = 'white',
                      max_words = 100, colormap = 'viridis').generate(' '.join(words))

plt.figure(figsize = (10, 5))
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis("off")
plt.show()
```

```python
from wordcloud import WordCloud


wordcloud = WordCloud(width = 800, height = 400, background_color = 'white',
                      max_words = 100, colormap = 'viridis', random_state = 42).generate(' '.join(words))

plt.figure(figsize = (10, 5))
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis("off")
plt.show()
```
